```python
import urllib.request
import time
import csv
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = 2017
months = 11
days = 1
```

```python
for year in range(years,2019):
    for m in range(months,13):       
        for d in range(days,32):
            #---------------------------------期貨資料下載 - --------------------------------------
            if m >= 10:
                if d < 10:
                    url = ('http://www.tse.com.tw/exchangeReport/MI_5MINS_INDEX?response=csv&date=' + str(year) + str(m) + '0' + str(d))
                else:
                    url = ('http://www.tse.com.tw/exchangeReport/MI_5MINS_INDEX?response=csv&date=' + str(year) + str(m) + str(d))
            else:
                if d < 10:
                    url = ("http://www.tse.com.tw/exchangeReport/MI_5MINS_INDEX?response=csv&date=" + str(year) + '0' + str(m) + '0' + str(d))
                else:
                    url = ('http://www.tse.com.tw/exchangeReport/MI_5MINS_INDEX?response=csv&date=' + str(year) + '0' + str(m) + str(d))
            try:
                f = urllib.request.urlretrieve(url,'C:/新期貨資料/' + str(year) + ' ' + str(m) + ' ' + str(d) + '.csv')
                time.sleep(2)
            except urllib.error.URLError:
                continue
            except FileExistsError:
                continue
```

```python
data_cut = open('C:/期貨cutdata/' + 'datamixsmall2' + '.csv','x')

for year in range(years,2018):
    for m in range(months,13):       
        for d in range(days,32):
            try:
                data = open('C:/新期貨資料/' + str(year) + ' ' + str(m) + ' ' + str(d) + '.csv','r')
            except FileNotFoundError:
                continue
            csvdata = csv.reader(data,delimiter=',', quotechar='"')
            
            csvdata_cut = csv.writer(data_cut,delimiter=',',quotechar='"',lineterminator='\r')

            flag = 0
            #取前兩筆資料 時間 價格
            for row in csvdata:
                
                if len(row) < 2:
                    continue
                     
                if row[0] == str('="09:59:55"'):
                    flag = 1
                row = row[1:2].replace(',','')
                if flag == 1:
                    row[0] = '="' + row[0] + '"'
                    csvdata_cut.writerow(row)
                else:
                    csvdata_cut.writerow(row)

                if row[0] == str('="09:59:55"'):
                    flag = 1
                print(repr(row))
```
